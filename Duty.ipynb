```python
import pandas as pd
import numpy as np
```

```python
def auto_count(months):
    
    def read(csv, weeks):
        month = pd.read_csv(csv)
        month.columns = month.iloc[0]
        month = month.iloc[::2]
        month.reset_index()
        return sum(month.values.tolist(), weeks)
    def count(weeks):
        for i in weeks:
            if pd.notna(i):
                if i[0] == "9":
                    day = i.split("\n")
                    for j in day:
                        duty[j[j.index(":") + 1:].lstrip().split(" ")[0]] += 1
                elif i[0] == "\n":
                    day = i.split("\n")[-1]
                    duty[day[day.rfind(":") + 1:].lstrip()] += 1
                elif i[0] == "L":
                    day = i.split("\n")
                    lunch[day[0][day[0].index(":") + 1:].lstrip().split(" ")[0]] += 1
                    duty[day[-1][day[-1].index("pm:") + 3:].lstrip().split(" ")[0]] += 1
                elif i[:3] == "Mot":
                    day = i.split("\n")
                    for j in day[3:]:
                        duty[j[j.index(":") + 1:].lstrip().split(" ")[0]] += 1
                elif i[:2] == "Co":
                    day = i.split("\n")
                    for j in day[3:]:
                        duty[j[j.index(":") + 1:].lstrip().split(" ")[0]] += 1
    
    def get_duty():
        duties = dict(duty)
        duties["Jenn"] += duties["Jennifer"]
        duties["Jenn"] += duties["Jen"]
        duties.pop("Jennifer")
        duties.pop("Jen")
        duties["JJ"] += duties["Llajayra"]
        duties.pop("Llajayra")
        duties["Alex"] += duties["Alexis"]
        duties.pop("Alexis")
        
        duties.pop("Emilio")
        duties.pop("")
        
        return duties
    
    def get_lunch():
        lunches = dict(lunch)
        lunches["Jenn"] += lunches["Jennifer"]
        lunches["Jenn"] += lunches["Jen"]
        lunches.pop("Jennifer")
        lunches.pop("Jen")
        lunches["JJ"] += lunches["Llajayra"]
        lunches.pop("Llajayra")
        lunches["Alex"] += lunches["Alexis"]
        lunches.pop("Alexis")
        
        lunches.pop("Emilio")
        lunches.pop("")
        return lunches
    
    names = ["Daniella", "Nicole", "Eden", "Maria", "Catherine", "Cecilia", "Jenn", "Jennifer", "Jen", "Savannah", 
             "Llajayra", "JJ", "Louie", "Amber", "Gabbey", "Angela", "Jung", "Ben", "Brigette", "Siri", "Kanishk", 
             "Anushka", "CJ", "George", "Emma", "Rachel", "Sophia", "Jade", "Jessie", "Alex", "Alexis", "Luzia", "Emilio", ""]
    duty = dict.fromkeys(names, 0)
    lunch = dict.fromkeys(names, 0)
    
    for i in months:
        weeks = []
        weeks = read(i, weeks)
        count(weeks)
    d = pd.DataFrame.from_dict(get_duty(), orient = "index")
    d = d.rename(columns={d.columns[0]: 'Duty'})
    l = pd.DataFrame.from_dict(get_lunch(), orient = "index")
    l = l.rename(columns={l.columns[0]: 'Lunch'})
    return pd.concat([d, l], axis = 1)
```

```python
months = ["September", "October", "November", "December", "January", "February", "March", "April", "May", "June"]
monthcsvs = []
for i in months:
    monthcsvs.append(f"2022-2023 Res Life Copy HA Duty_Office Shifts  - {i}.csv")
ds = auto_count(monthcsvs)
ds
```
